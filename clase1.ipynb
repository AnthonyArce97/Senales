```
import numpy as np
import matplotlib.pyplot as plt

# INGRESO
a= -5 # rango de tiempo
b= 5
dt= 0.1
sigma= 0 # s= 0+0j
omega= 0

# PROCEDIMIENTO
t = np.arange(a,b,dt)
s = complex(sigma,omega)
senal=np.exp(s*t)

# SALIDA - gr√°fica
plt.figure(1)
plt.plot(t,np.real(senal),label='Real')
plt.plot(t,np.imag(senal),label='Imginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma)+'+'+str(omega)+'j)')
plt.show()
```

```
import scipy.integrate as integrate 

w=1
a=0.1
b=0
c=2*np.pi
dte=0.005


te= np.arange(b,c,dte)
xe=np.exp(-a*te)


cuadradoxe= xe**2
energiaxe = integrate.simps(cuadradoxe,te)

#Salida

print('La energia de xe es: ', energiaxe)

#Grafica

plt.plot(te,xe,label='xe')
plt.fill_between (te,0,cuadradoxe,color='Lightgreen')
plt.plot(te,cuadradoxe,'g--',label='xe^2')
plt.xlabel('t')
plt.legend()
plt.show()
```

```
w=1
a=0
b= 2*np.pi
dte=0.005
Q=0
A=1


T= 2*np.pi/w
te= np.arange(a,b,dte)
xe=A*np.cos(w*te+Q)

cuadradoxe = xe**2
energiaxe = integrate.simps(cuadradoxe,te)
#Potencia en un periodo
potenciaxe = (1/T)*energiaxe
#Salida
print('la potencia de xe es', potenciaxe)

#grafica
plt.fill_between(te,0,xe,color='green')
plt.plot(te,xe,label='xe')
plt.fill_between(te,0,cuadradoxe,color='lightgreen')
plt.plot(te,cuadradoxe,'g--',label='xe^2')
plt.xlabel('t')
plt.legend()
plt.show()
4
```
